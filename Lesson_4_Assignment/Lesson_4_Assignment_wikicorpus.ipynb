```python
from hanziconv import HanziConv
from gensim.corpora import WikiCorpus
import jieba
```

```python
# xmltxt
def my_function():
    space = " "
    i = 0
    l = []
    zhwiki_name = "C:\\Users\\ywan3\\Desktop\\NLP\\Data\\zhwiki-20190720-pages-articles-multistream.xml.bz2"
    f = open('./reduce_zhwiki.txt', 'w')
    wiki = WikiCorpus(zhwiki_name, dictionary={}) # xml
    for text in wiki.get_texts():
        for temp_sentence in text:
            temp_sentence = HanziConv.toSimplified(temp_sentence) # 
            seg_list = list(jieba.cut(temp_sentence))
            for temp_term in seg_list:
                l.append(temp_term)
        f.write(space.join(l) + '\n')
        l = []
        i = i + 1

        if (i %200 == 0):
            print('Saved ' + str(i) + ' articles')
    f.close()

if __name__ == '__main__':
    my_function()
```
